```python
import yaml
import json

```

```python
# Read data from the .yaml file
with open('demagog_Babis.yaml', 'r') as file:
    data = yaml.safe_load(file)

# Initialize an empty list to store the formatted messages
formatted_messages = []

# Iterate over the data items and create the formatted messages
for item in data:
    msg = []
    msg.append({"role": "system", "content": "Jsi " + item['name'] + " , esk politik, bval premir a miliard."})
    msg.append({"role": "user", "content": item['truth']})
    msg.append({"role": "assistant", "content": item['quote']})
    formatted_msg = {"messages": msg}
    formatted_messages.append(formatted_msg)

# Convert the formatted data to a JSON string with brackets
formatted_json = json.dumps(formatted_messages, indent=2)

# Print the formatted JSON
print(formatted_json)
```

```python
json_data_list = []

# Iterate through each item and convert it to JSON
for item in formatted_messages:
    json_data_list.append(item)

# Create a dictionary to store all the items

# Save the data to a JSON file
output_file = 'output.json'
with open(output_file, 'w') as outfile:
    json.dump(formatted_messages, outfile, ensure_ascii=False, indent=2)

print(f"Data saved to {output_file}")
```

```python
input_file = 'output.json'
with open(input_file, 'r') as infile:
    data = json.load(infile)

# Process the data to remove parentheses and "pozn. Demagog.cz" parts
for item in data:
    for message in item['messages']:
        message['content'] = message['content'].replace('(', '').replace(')', '').replace('pozn. Demagog.cz', '')

# Save the modified data back to the same JSON file
with open(input_file, 'w') as outfile:
    json.dump(data, outfile, ensure_ascii=False, indent=2)

print(f"Data saved to {input_file} with parentheses and 'pozn. Demagog.cz' parts removed.")
```

```python
# Load data from the input JSON file
input_file = 'output.json'
with open(input_file, 'r') as infile:
    data = json.load(infile)

# Modify the data as needed (for example, removing extra spaces and trailing periods)
modified_data = []

for item in data:
    messages = item['messages']
    for message in messages:
        message['content'] = message['content']
    modified_data.append({"messages": messages})

# Save the modified data to an output JSON file
output_file = 'outputModified.json'
with open(output_file, 'w') as outfile:
    for item in modified_data:
        json.dump(item, outfile, ensure_ascii=False, separators=(', ', ': '), indent=None)
        outfile.write('\n')  # Add a newline character after each message

print(f"Modified data saved to {output_file} with spaces after commas and colons")

```
