```python
import requests
from bs4 import BeautifulSoup
import copy
import yaml
import os
import re
import pandas as pd

#function for scraping webpage demagog.cz for politicians quotes-> right now it is set to Andrej Babis
#politician = "Andrej Babiš"
URL_demagog = 'https://demagog.cz/politici/andrej-babis-183'
filename = "./data/babis/demagog_Babis.yaml"

#URL_vadne = 'https://demagog.cz/politici/andrej-babis-183?page=48'
```

```python
def get_url_data(URL, politician):
    
    data = []

    page = requests.get(URL)
    if page.status_code != 200:
        raise ValueError(f"Failed to fetch the webpage. HTTP Status Code: {page.status_code}")
    html_content = page.text
    soup = BeautifulSoup(html_content, "html.parser")
    
    class_pattern_states = re.compile(r"fs-5 text-uppercase fs-600 text-(\w+)")
    
    quotes_all = soup.find_all('span', class_="fs-6 position-relative")
    articles_all = soup.find_all('article', class_='mb-10 s-statement')
    states_all = soup.find_all(attrs={"class": class_pattern_states})
    truths = []
    
    for article in articles_all:
        truth_current = ""
        short = article.find('div', class_='content fs-6')
        long = article.find_all('p')
        if short:
            truth_current = short.get_text(strip=True)
        else:
            for element in long:
                truth_current += element.get_text(strip=True)
        truths.append(truth_current)

    '''
    print(len(quotes_all))
    print(len(states_all))
    print(truths)
    
    for i in range(9):
        print(quotes_all[i].text + ", " + states_all[i].string + ", " + truths_all[i].string + "\n")
    '''
    if(len(quotes_all) != len(truths) != len(states_all)):
        print("Data lengths on URL: " + URL + " are not equal and do not correspond to each other.")
    
    for idx in range(len(quotes_all)):
        quote_current = quotes_all[idx].text
        quote_current = quote_current.replace('\xa0', ' ').replace('\n',' ').strip()
        #print(quote_current)

        state_current = states_all[idx].string
        state_current = state_current.strip().replace('\n',' ').strip()
        #print(state_current)
        truth_current = truths[idx].replace('\xa0', ' ').replace('\n',' ').strip()
        #print(truth_current)

        data_current = {"name": politician, "quote": quote_current, "state": state_current, "truth": truth_current}
        data.append(data_current)
  
    return data
```

```python
def url_cycle(URL, politician, filename):

    data_all = get_url_data(URL_demagog,"Andrej Babiš")
    print("The 1. URL(" + URL + ") has been scraped.")
    
    for idx in range(105):
        i = idx + 2
        URL_current = URL + "?page=" + str(i)
        data_current = get_url_data(URL_current, politician)
        data_all.append(data_current)
        print("The " + str(i) + ". URL(" + URL_current + ") has been scraped.") 

    with open(filename, "w", encoding="utf-8") as file:
        yaml.dump(data_all, file, default_flow_style=False, allow_unicode=True)
    
    #df = pd.DataFrame(data_all)
    #df.to_csv(filename, index=False)

    print("All data saved to file: " + filename + ". ")

    return 
```

```python
url_cycle(URL_demagog, "Andrej Babiš", filename)
#print(get_url_data(URL_vadne,"Andrej Babiš"))
```
